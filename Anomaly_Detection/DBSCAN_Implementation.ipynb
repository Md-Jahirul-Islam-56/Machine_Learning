```python
from sklearn.cluster import DBSCAN
from sklearn.datasets import make_moons
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
X,y=make_moons(n_samples=250,noise=0.1)
```

```python
X
```

```python
plt.scatter(X[:,0],X[:,1])
```

```python
##feature scaling(Standard Scaling)
from sklearn.preprocessing import StandardScaler
scaler=StandardScaler()
```

```python
X_scaled=scaler.fit_transform(X)
```

```python
X_scaled
```

```python
dbscan=DBSCAN(eps=0.15)
```

```python
dbscan.fit(X_scaled)
```

```python
dbscan.labels_
```

```python
#Anomaly Detection
labels = dbscan.fit_predict(X)
anomalies = X[labels == -1]
normal_points = X[labels != -1]
```

```python
plt.scatter(normal_points[:, 0], normal_points[:, 1], c='blue', label='Normal')
plt.scatter(anomalies[:, 0], anomalies[:, 1], c='red', label='Anomaly')
plt.legend()
plt.title("DBSCAN Anomaly Detection")
plt.show()
```
